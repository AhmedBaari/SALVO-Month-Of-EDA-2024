```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('shopping_trends.csv')
```

```python
df
```

```python
df=df.set_index('Customer ID')
```

```python
df.drop_duplicates()
```

```python
df = df.drop(columns = ['Previous Purchases','Promo Code Used','Discount Applied','Subscription Status'] )
```

```python
df
```

```python
df.loc[df['Frequency of Purchases'] == 'Annually', 'Frequency of Purchases'] = 1
df.loc[df['Frequency of Purchases'] == 'Quarterly', 'Frequency of Purchases'] = 4
df.loc[df['Frequency of Purchases'] == 'Monthly', 'Frequency of Purchases'] = 12
df.loc[df['Frequency of Purchases'] == 'Weekly', 'Frequency of Purchases'] = 52
df.loc[df['Frequency of Purchases'] == 'Fortnightly', 'Frequency of Purchases'] = 26
df.loc[df['Frequency of Purchases'] == 'Bi-Weekly', 'Frequency of Purchases'] = 104
df.loc[df['Frequency of Purchases'] == 'Every 3 Months', 'Frequency of Purchases'] = 4
```

```python
df
```

```python
df['Age_Category_Interact'] = df['Age'].astype(str) + '_' + df['Category']
```

```python
df
```

```python
sum_by_category = df.groupby('Category')['Frequency of Purchases'].sum().reset_index()
plt.bar(sum_by_category['Category'], sum_by_category['Frequency of Purchases'])
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Category VS Frequency of Purchase')
plt.show()
```
